Yukarida goruldugu gibi esit araliklarla tek duzlem uzerinde veriler bulunmaktadir. Ust kisminda da hangi sinifa ait olduklari belirtilmektedir. Biz bize gelen yeni datanin (orneklerde 1.6'yi kullandik) hangi sinifa ait oldugunu sorarsak?  

```python
input=[[0],[1],[2],[3]]
output=[0,0,1,1]
from sklearn.neighbors import KNeighborsClassifier

my_blackbox=KNeighborsClassifier(n_neighbors=1,weights='uniform')
my_blackbox.fit(input,output)

my_blackbox.predict([[1.6]]) #1.6 un nerde oldugunu tahmin edecek

print(my_blackbox.predict([[1.6]]))

#bunu satirin aynisini irisdata icin yazsaydik nasil olurdu [W5_1de]
#predict([[5.1,3.5,1.4,0.2]]) seklinde olurdu
```

**KNN ALGORITHM** <br>
n_neighbors= k <br>
* k=1 icin komsu olarak 1&2 ayni uzaklikta oldugu icin birini secmesi gerekir. Bilgisayar en yakin olarak (0-1-2.. diye sayiyor diye dusunup ilk soyledigini daha yakin sayar) 1'i secer.
* k=2 icin komsu olarak 1&2'yi secer. Hangisi daha yakin sorusunun cevabi %50 %50 gelir ve 1'i secer.
* k=3 icin komsu olarak 0&1&2'yi secip en yakin olarak yine 1'i sececek.


```python
#n_neighbors=2 yaparsak

input=[[0],[1],[2],[3]]
output=[0,0,1,1]

from sklearn.neighbors import KNeighborsClassifier

my_blackbox=KNeighborsClassifier(n_neighbors=2,weights='uniform')
my_blackbox.fit(input,output)


print(my_blackbox.predict([[1.6]]))
print(my_blackbox.predict_proba([[1.6]]))
```

***.predict_proba()***

*my_blackbox_d.predict_proba([ [1.6] ])*

* proba= olasilik, siniflara ait olma ihtimallerini verir (tek duzlemde dusununce siniflara olan uzakligini veriyor)
* Benim iki sinifim vardi c1=0 ve c2=1 yani bana bu iki sinifa olan uzakligi vericek\ benim onlara ait olma ihtimalimi
* Bunu print ettigimizde cikti bize kac tane sinif varsa hepsi icin icinde olma ihtimalini verir <br>-->
 *print(my_blackbox_d.predict_proba([[1.6]]))*   --> [[0.4   0.6]]

```python
#n_neighbors=2 & weights='distance' yaparsak

input=[[0],[1],[2],[3]]
output=[0,0,1,1]
from sklearn.neighbors import KNeighborsClassifier

my_blackbox_d=KNeighborsClassifier(n_neighbors=2,weights='distance')
my_blackbox_d.fit(input,output)


print(my_blackbox_d.predict([[1.6]]))
print(my_blackbox_d.predict_proba([[1.6]]))
```

```python
#n_neighbors=3 & weights='distance' yaparsak ve bu sefer 1.9'un sinifini sorarsak

input=[[0],[1],[2],[3]]
output=[0,0,1,1]

from sklearn.neighbors import KNeighborsClassifier

my_blackbox_d=KNeighborsClassifier(n_neighbors=3,weights='distance')
my_blackbox_d.fit(input,output)


print(my_blackbox_d.predict([[1.9]]))
print(my_blackbox_d.predict_proba([[1.9]]))
```

# ***Sinif sayisini 3'e cikarirsam yani c3=2'yi eklersem***

```python
#n_neighbors=2 & weights='uniform' yaparsak ve 1.99'un sinifini sorarsak

input=[[0],[1],[2],[3]]
output=[0,0,1,2]

from sklearn.neighbors import KNeighborsClassifier

my_blackbox=KNeighborsClassifier(n_neighbors=2,weights='uniform')
my_blackbox.fit(input,output)


print(my_blackbox.predict([[1.99]]))
print(my_blackbox.predict_proba([[1.99]]))
```

```python
#n_neighbors=2 & weights='distance' yaparsak ve 1.99'un sinifini sorarsak

my_blackbox_d=KNeighborsClassifier(n_neighbors=2,weights='distance')
my_blackbox_d.fit(input,output)

print(my_blackbox_d.predict([[1.99]]))
print(my_blackbox_d.predict_proba([[1.99]]))
```

```python
#n_neighbors=3 & weights='distance' yaparsak ve 1.99'un sinifini sorarsak

my_blackbox_d=KNeighborsClassifier(n_neighbors=3,weights='distance')
my_blackbox_d.fit(input,output)

print(my_blackbox_d.predict([[1.99]]))
print(my_blackbox_d.predict_proba([[1.99]]))
```
