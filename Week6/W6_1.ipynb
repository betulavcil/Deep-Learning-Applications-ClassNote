```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as myplotter

# Lets read the data
myirisdata=load_iris()
#print(myirisdata)
X=myirisdata['data']
Y=myirisdata['target']


mymachine=DecisionTreeClassifier(criterion='gini',max_depth=20)
mymachine.fit(X,Y)
#sadece ekranima duzgun gelsin diye isim ekleyecegim, kisisel tercih
feature_name=['SP-L','SP-W','PT-L','PT-W']

#burasi kalip hep ayni seyi kullanicaz
# myplotter.figure()
# resmi buyutmek icin figure(figsize=(17,17)) ne kadar buyutmek istersen sayiyi ona gore degistir
myplotter.figure(figsize=(17,17))
plot_tree(mymachine,filled=True)
myplotter.show()
```

```python
#entropy kullanicaz
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as myplotter

# Lets read the data
myirisdata=load_iris()
X=myirisdata['data']
Y=myirisdata['target']

# criterion = entropy yaptik
mymachine=DecisionTreeClassifier(criterion='entropy',max_depth=20)
mymachine.fit(X,Y)

myplotter.figure(figsize=(17,17))
plot_tree(mymachine,filled=True)
myplotter.show()
```
