```python
# iris dataset
my_url_address="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

my_data_set_x= pd.read_csv(my_url_address)
print(my_data_set_x)
```

***pd.read_csv("url"):***   

*   pd= dosya islemleri
*   .read= okuma islemleri
* _csv: coma seperated value : virgul ile ayrilmis deger

Bir dosyadaki degeri okuyup virgulden virgule ayirir. <br>

Eger dosya virgul ile ayrilmamissa bunu kullanman bir sey ifade etmez bu yuzden once dataseti indirip notepad da acip datanin nasil kaydedildigine bakmalisin (bazilari bosluk ile ayrilmis olarak yaziliyor)
<br>

(iris dataset virgul ile ayrilmis iken heart diseases dataset bosluk ile ayrilmisdir bu yuzden de csv burada ise yaramiyor)


```python
# ciktinin columnlarina isim vermek istersek
name_of_columns=["x1","x2","x3","x4","Label"]
my_data_set_x=pd.read_csv(my_url_address,names=name_of_columns)
print(my_data_set_x)
# x'ler input, Label outputtur
```

Ben datasetimi UCI'dan bulup (istersem download edip datasetimi inceleyebilirim)  

*   Url'sini buraya tanimlayip -> *my_url_address="link"*
*   Bir degiskene bunu oku diye atariz --> *my_data_set_x=pd.read_csv(my_url_address)*
* Print ettigimizde bize bu dataseti gosterir, artik dataseti burada inceleyebiliriz.
<br>

* Eger biz bu datasetin sutunlarina isim vermek istersek (incelememizi ve gorunumu kolaylastirmak adina) yapmamiz gerekenler:  

> 1.   sutunlara verilecek isimlerin oldugu bir array olustur --> *name_of_columns=["x1","x2","x3","x4","label"]*
2.  Datayi oku diye degiskene atarken isim arrayini de *csv()* fonksiyonun icine parametre olarak atariz-->  *my_data_set_x=pd.read_csv(my_url_address,names=name_of_columns)*
