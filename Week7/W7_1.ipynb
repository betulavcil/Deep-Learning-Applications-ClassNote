```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.datasets import load_iris

mydata=load_iris()
X=mydata['data']
Y=mydata['target']

print(X,Y) # nasıl bir data oldugunu goruntulemek icin
```

```python
#Normalization Skalers
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
#This object definitions
#normalize yapmak icin standart & minmax olusturmamiz lazim once

mystandardscaler= StandardScaler()
mystandardscaler.fit(X)

minmaxscaler=MinMaxScaler()
minmaxscaler.fit(X)

X1=mystandardscaler.transform(X)
X2=minmaxscaler.transform(X)
#sadece inputlar normalize edilir outputlar normalize edilmez

#print(Y)
```

```python
#testing training ayiriyoruz
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
X1_train,X1_test,Y1_train,Y1_test=train_test_split(X1,Y,test_size=0.25)
X2_train,X2_test,Y2_train,Y2_test=train_test_split(X2,Y,test_size=0.25)
#y yi yani outputu normalize etmiyorum
```

# ** WINE DATASET**

```python
#iris data bunun icin iyi bir ornek degil farkli dataset ile deniyoruz
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

mydatawine=load_wine()

A=mydata['data']
B=mydata['target']



mstandartscaler= StandardScaler()
mstandartscaler.fit(A)

minmaxscale=MinMaxScaler()
minmaxscale.fit(A)

A1=mstandartscaler.transform(A)
A2=minmaxscale.transform(A)

A_train,A_test,B_train,B_test=train_test_split(A,B,test_size=0.25)
A1_train,A1_test,B1_train,B1_test=train_test_split(A1,B,test_size=0.25)
A2_train,A2_test,B2_train,B2_test=train_test_split(A2,B,test_size=0.25)


mknn=KNeighborsClassifier(n_neighbors=5,weights='uniform')
mknn1=KNeighborsClassifier(n_neighbors=5,weights='uniform')
mknn2=KNeighborsClassifier(n_neighbors=5,weights='uniform')

mknn.fit(A_train,B_train)
mknn1.fit(A1_train,B_train)
mknn2.fit(A2_train,B_train)

B_test_calculated=mknn.predict(A_test)
B1_test_calculated=mknn1.predict(A1_test)
B2_test_calculated=mknn2.predict(A2_test)

#lets print the results
print('Original Data')
print(confusion_matrix(B_test,B_test_calculated))
print(classification_report(B_test,B_test_calculated))
print('Standart Scaler')
print(confusion_matrix(B1_test,B1_test_calculated))
print(classification_report(B1_test,B1_test_calculated))
print('Minmax Scaler')
print(confusion_matrix(B2_test,B2_test_calculated))
print(classification_report(B2_test,B2_test_calculated))
```
