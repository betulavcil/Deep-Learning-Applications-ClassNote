```python
#Change from matrix to array of dimension 28x28 to array of dimension 784
dimData=np.prod(train_images.shape[1:])
train_data=train_images.reshape(train_images.shape[0],dimData)
test_data=test_images.reshape(test_images.shape[0],dimData)
#Elimdeki datayi flatting yapmayi sagliyor
```

```python
#Change to float datatype

train_data=train_data.astype('float32')
test_data=test_data.astype('float32')

#Scale the data to lie between 0 to 1
train_data /= 255
test_data /= 255
```

```python
#The classification part
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

model.add(Dense(512, activation='tanh', input_shape=(dimData,)))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
#10 classes, 10 process elemani olur

model.add(Dense(10, activation='softmax'))
#son katman--> softmax, en cok neye benziyorsa bana onun hakkini vermesi icin
```
