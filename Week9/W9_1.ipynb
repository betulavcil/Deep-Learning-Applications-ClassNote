```python
#Lets import Keras and Tensorflow
import tensorflow as tf
from tensorflow import keras

#ANN and DeepLearning tools
from tensorflow.keras.layers import Conv2D, MaxPooling2D,Dropout,Flatten,Dense
#Conv2D, MaxPooling2D,Dropout,Flatten convolution neural network (CNN) icin bugun bunlarla isimiz yok
#we need Dense layer for simple ANN

import numpy as np

#Regression example - Boston House Pricing
#google yazip kaggledan datasete bakabilirsin
from keras.datasets import boston_housing
(X_train,Y_train),(X_test,Y_test)=boston_housing.load_data()
#bu ikisi keras'a ozel bir kalip
```

```python
X_train.shape
#Shape boyutu verir
#outputa gore : 404 satir 13 atribute var
```

```python
print(X_train)
# outputa bak datalar normalize edilmis
```

```python
#Sequential : sirali sekilde aglardan olusana seq diyor (biz sadece seq olusturucaz)
from keras.models import Sequential
#For simple ANN needs use Sequential Model

my_ann_model=Sequential()
#yapay sinir aglarinda calisacaksan ilk once bir modele ihtiyacin var
#ilk once neye karar vericem model olustururken? --->input
```

```python
print(X_train[1])

print(X_train.shape[1])  #bu direkt kac atribute oldugu verir (yani 13)
```

```python
from keras.engine import input_layer

print('dataset size', X_train.shape)
#yukaridan aldigimiz outputlara gore 13 process elemana ihtiyacim var

input_layer_size=X_train.shape[1]

#Define the model with two layers

my_ann_model.add(Dense(26,input_shape=(input_layer_size,),activation='tanh'))
my_ann_model.add(Dense(50,activation='tanh'))


my_ann_model.compile(optimizer='rmsprop',loss='mse',metrics=['mse','mae'])

my_ann_model.fit(X_train,Y_train,batch_size=5,epochs=100)
my_ann_model.summary()
my_ann_model.evaluate(X_test,Y_test)

# bunun sonucu duzgun cikmadi ogrenemedi bu bi sanki
```
